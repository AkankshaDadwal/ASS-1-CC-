```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='b', linestyle='--')
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='r', linestyle='-')

plt.title(f"Plots of y = MxÂ² and y = Msin(x) for M = {M}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Q.2 Create a dataset of five subjects and scores.
ï· Convert it into a Pandas DataFrame.
ï· Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'Subjects': ['Math', 'Science', 'English', 'History', 'Art'],
        'Scores': [85, 92, 78, 88, 95]}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='Subjects', y='Scores', data=df, palette='husl')

for index, row in enumerate(df['Scores']):
    barplot.text(index, row + 1, str(row), ha='center', fontsize=12, fontweight='bold')

plt.title("Student Scores by Subject")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 100)

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

 Create a 2x2 subplot layout displaying:

o Line plot showing cumulative sum.
o Scatter plot with random noise.

 Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 123456  # Replace with your actual roll number
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), color='b', marker='o', linestyle='-')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

axes[0, 1].scatter(range(50), data, color='r', alpha=0.7)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Values")
axes[0, 1].grid(True)

fig.delaxes(axes[1, 0])
fig.delaxes(axes[1, 1])

plt.tight_layout()
plt.show()
```

Apply âseabornâ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Total_Profit': [20000, 25000, 27000, 30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000],
    'Product_A_Sales': [500, 600, 750, 800, 850, 900, 950, 1000, 1100, 1150, 1200, 1300],
    'Product_B_Sales': [450, 500, 550, 600, 650, 700, 750, 850, 900, 950, 1000, 1100]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.lineplot(x='Month', y='Total_Profit', data=df, marker='o', color='b')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
sns.lineplot(x='Month', y='Product_A_Sales', data=df, marker='o', label="Product A", color='g')
sns.lineplot(x='Month', y='Product_B_Sales', data=df, marker='s', label="Product B", color='r')
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

df_melted = df.melt(id_vars=['Month'], var_name="Feature", value_name="Value")

plt.figure(figsize=(8, 5))
sns.barplot(x='Feature', y='Value', data=df_melted, palette='muted')
plt.title("Bar Chart of Features")
plt.xlabel("Features")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
